```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics  import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Dataset_1/train.csv')
testing_set = pd.read_csv('Dataset_1/test.csv')
```

```python
df.dropna(inplace=True)
```

```python
X = df[['x']]
y = df[['y']]
X_test = testing_set[['x']]
y_test = testing_set[['y']]
```

```python
#SKLEARN
model = LinearRegression()

model.fit(X, y)
model_predictions = model.predict(X_test)
mean_squared_error(y_test, model_predictions)
```

```python
plt.figure(figsize = (5, 4))
plt.scatter(X_test, testing_set[['y']],color = 'black',label='Test.csv data')
plt.plot(X_test,model_predictions, color = 'blue', label = 'Sklearn model')
plt.title('Scikit-learn LinearRegression implementation ')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
